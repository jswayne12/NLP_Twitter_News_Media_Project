```python
%sql SELECT tweet_id,political_alignment, news_sources_type, length_of_tweet, day_of_week, NULLIF (overall_sentiment, 0.0000) AS overall_sentiment,\
NULLIF (subjectivity, 0.000) as subjectivity from $sql_table \
WHERE length_of_tweet >= 10
```
